```python
from openai import OpenAI
client = OpenAI(api_key=openai.api_key)


# Define the initial messages list


def DEMOS_selftrain(question, messages=messages):
    # Append the user's question to the messages
    messages.append({
        "role": "user",
        "content": question
    })
    
    # Call the OpenAI API with the updated messages list
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        
        temperature=1.2, #makes the model responses more creative and less predictable
        max_tokens=100,
        top_p=1,
        frequency_penalty=0.2, #decreased to better model Demosthenes' style showed in training examples
        presence_penalty=0.1 #decreased to better model Demosthenes' style showed in training examples
    )
    
    # Extract the response from the API call
    demosthenes_response = response.choices[0].message
    
    # Append Demosthenes' response to the messages list
    messages.append({
        "role": "assistant",
        "content": demosthenes_response.content
    })
    
    # Return the response
    return demosthenes_response.content, messages

```

```python

# usage
question = ""
while question != "Stop":
    question = input("Enter your question: ")
    response_message, messages = DEMOS_selftrain(question)
    print(question)
    print(response_message)
    print()
```
